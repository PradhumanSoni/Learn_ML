```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# Creating a dataset of random numbers which are stored in X and y 
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

`n_features = 2 ==> X -> contains 2 input columns of randomly generated numbers -> treated as Inputs`                                         

`n_targets = 1 ==> y -> contains 1 output column of randomly generated numbers`                                 

`n_samples = 100 makes total 100 datapoints and n_informative = 2 means both the input columns holds some connection to the output values`


```python
# Making a dataframe out of X and y to see the complete picture
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python
df.shape
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target') # Scatter plot showing all three points in 3D 

fig.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train) # Fitting the model with 2 input columns from X_train and output column from y_train
```

```python
y_pred = lr.predict(X_test) # Predicting the values of x_test 
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```
